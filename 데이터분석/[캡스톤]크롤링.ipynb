```python
import time
from tqdm import trange
```

```python
# 1. '확진자 현황(40001번~현재)' 탭의 page_no 페이지에 존재하는 데이터 요청
def get_seoul_covid19_40001_current(page_no):
	
    start_no = (page_no - 1) * 100
    
    url = f"https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax.php?draw={page_no}"
    url = f"{url}&order%5B0%5D%5Bdir%5D=desc&start={start_no}&length=100"
    
    response = requests.get(url)
    data_json = response.json()
    
    return data_json
    
    
# 2. '확진자 현황(1번 ~ 40000번)' 탭의 page_no 페이지에 존재하는 데이터 요청
def get_seoul_covid19_1_40000(page_no):
    

    start_no = (page_no - 1) * 100
    
    url = f"https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax_pre.php?draw={page_no}"
    url = f"{url}&order%5B0%5D%5Bdir%5D=desc&start={start_no}&length=100"
        
    response = requests.get(url)
    data_json = response.json()
    
    return data_json
```

```python
# 1) '확진자 현황(40001번 ~ 현재)' 탭의 전체 페이지에 존재하는 데이터
def get_multi_page_list_40001_current(start_page, end_page=526):
	
    page_list = []
    for page_no in trange(start_page, end_page + 1):
      one_page_data_json = get_seoul_covid19_40001_current(page_no)
        
      if len(one_page_data_json["data"]) > 0:
        one_page_df = pd.DataFrame(one_page_data_json["data"])
        page_list.append(one_page_df)
        time.sleep(0.5)
      else:
        return page_list
     
    return page_list
     
     
# 2) '확진자 현황(1번 ~ 40000번)' 탭의 전체 페이지에 존재하는 데이터
def get_multi_page_list_1_40000(start_page, end_page=526):

  page_list = []
  for page_no in trange(start_page, end_page + 1):
    one_page_data_json = get_seoul_covid19_1_40000(page_no)     
    if len(one_page_data_json["data"]) > 0:
      one_page_df = pd.DataFrame(one_page_data_json["data"])
      page_list.append(one_page_df)
      time.sleep(0.5)
    else:
      return page_list
    
  return page_list
```

```python
# 1. '확진자 현황(40001번 ~ 현재)'
start_page = 1
end_page = 526

page_list = get_multi_page_list_40001_current(start_page, end_page)
df_all_40001_current = pd.concat(page_list)

# pd.read_html(url)로 읽어온 table[5]의 칼럼명 리스트로 변환하여 df_all_40001_current에 적용
df_all_40001_current.columns = table[5].columns.tolist()
df_all_40001_current.head()
```

```python
start_page = 1
end_page = 400

page_list = get_multi_page_list_1_40000(start_page, end_page)
df_all_1_40000 = pd.concat(page_list)

df_all_1_40000.columns = table[5].columns.tolist()
df_all_1_40000.head()
```

```python
df_all = pd.concat([df_all_40001_current, df_all_1_40000])
df_all
```

```python
import re

def extract_num(num_string):
  if type(num_string) == str:
    num_string = num_string.replace("corona19", "")
    num = re.sub("[^0-9]","",num_string)
    num = int(num)
    return num
  else:
    return num_string


df_all["연번"] = df_all["연번"].map(extract_num)
df_all
```

```python
def extract_korean(origin_text):
  subtract_text = re.sub("[^가-힣]", "", origin_text) # 한글이 아닌 문자 제거
  return subtract_text

df_all["퇴원현황"] = df_all["퇴원현황"].map(extract_korean)
df_all.loc[df_all["퇴원현황"].isin(['']), "퇴원현황"] = np.nan
df_all
```
